```python
#Dropping columns that do not fit the research problem
df.drop(['Displayed comment', 'ID', 'Geo point','Address', 'Cars'], axis = 1, inplace = True)
df
```

```python
df.info()
```

```python
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018

# select city == Paris

df_paris = df[df['City'] =='Paris'] 

df_paris.head()
```

```python
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris 
df1 = df_paris.sort_values(['Bluecar counter', 'hour'], ascending=[False, False])
df1.head()
```

```python
# What is the most popular hour for returning cars?
df_paris['hour'].value_counts()
```

```python
# What station is the most popular Overall?
df_paris[(df_paris['Kind'] == 'STATION') & (df_paris['Status'] == 'ok')].groupby('Public name')['Public name'].count().sort_values(ascending= False)
```

```python
# What station is the most popular at the most popular picking hour?
popular = df_paris.sort_values(by=['hour'], ascending=False)
print(popular.head())
```

```python
#  What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code?

 # Overall?
df_paris[(df_paris['Kind'] == 'STATION') & (df_paris['Status'] == 'ok')].groupby('Postal code')['Postal code'].count().sort_values(ascending= False)
```

```python
#  What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code?

# At the most popular picking hour?
df_paris[(df_paris['Kind'] == 'STATION') & (df_paris['Status'] == 'ok') & (df_paris['hour'] == 17)].groupby('Postal code')['Postal code'].count().sort_values(ascending= False)
```
